```python
import duckdb
import pandas as pd
# Reading the cleaned credit data from a CSV file saved in the notebook 1
Credit_data = pd.read_csv('Credit_data_cleaned_SQL.csv')
```

```python
Credit_data.columns
```

```python
conn = duckdb.connect()
conn.register('Credit_data',Credit_data)
```

### 1. Grouping the customers based on their income type to find the average of their annual income.

```python
query = """ Select Type_Income, round(avg(Annual_income),2) as Average_Annual_income
            from Credit_data
            group by Type_Income
        """
conn.execute(query).fetchdf()
```

### 2. Finding the female owners of cars and property.

```python
query = """ Select Ind_ID, Gender, Car_Owner, Propert_Owner
            From Credit_data
            Where Gender = 'F' and Car_Owner = 'Y' and Propert_Owner = 'Y'
            Group By Ind_ID, Gender, Car_Owner, Propert_Owner
        """
print("ID's of the female owners of cars and property")
conn.execute(query).fetchdf()
```

### 3. Finding the male customers who are staying with their families.

```python
query1 = """ Select Gender, Housing_type, Count(*) as Total_Count
            From Credit_data
            Where Gender = 'M' and Housing_type = 'With parents'
            Group By Gender, Housing_type
        """
print("Total number of male customers who are staying with their families")
conn.execute(query1).fetchdf()
```

```python
query2 = """ Select Ind_ID, Gender, Housing_type
            From Credit_data
            Where Gender = 'M' and Housing_type = 'With parents'
            Group By Ind_ID, Gender, Housing_type
        """
print("ID's of male customers who are staying with their families")
conn.execute(query2).fetchdf()
```

### 4. Listing the top five people with the highest income.

```python
query = """ Select *
            From Credit_data
            Order by Annual_Income desc
            Limit 5
        """
print("Details of  the top five people having the highest income")
conn.execute(query).fetchdf()
```

### 5. Finding out how many married people have bad credit?

```python
query = """ Select Marital_status, Label, Count(*) as Bad_Credit
            From Credit_data
            Where Marital_status = 'Married' and Label = 'Rejected'
            Group By Marital_status, Label
        """
conn.execute(query).fetchdf()
```

- Looking at the results above 70 Married People have bad credit

### 6. What is the highest education level, and what is the total count?

```python
query = """ Select Education, Count(*) as Total_Count
            From Credit_data
            Group By Education
            Order By Education
        """
conn.execute(query).fetchdf()
```

- Looking at the results above Higher education is the highest education level with 305 total count.

### 7. Between married males and females, who has more bad credit?

```python
# SQL query to determine who, between married males and females, has more bad credit
query = """ Select Gender, Marital_Status, Label, Count(*) as Bad_Credit
            From Credit_data
            Where Marital_Status = 'Married' and Label = 'Rejected'
            Group By Gender, Marital_Status, Label
        """
conn.execute(query).fetchdf()
```

- Looking at the results above More Married Male have bad credit
